### 1. Linear Classifiers (logistic regression and GDA)

(a) Assume there is only one example in the training set. WLOG, it can be easily generalized to $m$ examples using similar steps.

\begin{equation*}
J(\theta) = -y\log(h_\theta(x))+(1-y)\log(1-h_\theta(x))
\end{equation*}
Then 
\begin{align*}
\frac{\partial J(\theta)}{\partial\theta_k\theta_j} &= -\frac{\partial}{\partial\theta_k}(y-g(\theta^Tx))x_j\\
&= g(\theta^Tx)(1-g(\theta^Tx))x_kx_j
\end{align*}
Thus 
\begin{align*}
z^THz &= g(\theta^Tx)(1-g(\theta^Tx))
\begin{bmatrix}
z_0 & z_1 & \cdots z_n
\end{bmatrix}
\begin{bmatrix}
x_0x_0 & x_0x_1 & \cdots & x_0x_n \\
x_1x_0 & x_1^2 & \cdots & x_1x_n \\ 
\vdots & & & \vdots\\
x_nx_0 & x_nx_1 & \cdots & x_nx_n
\end{bmatrix}
\begin{bmatrix}
z_0 \\ 
z_1 \\ 
\vdots \\ 
z_n
\end{bmatrix}\\
&= g(\theta^Tx)(1-g(\theta^Tx))\sum_i\sum_j z_ix_ix_jz_j \\
&= g(\theta^Tx)(1-g(\theta^Tx))(x^Tz)^2\geq 0
\end{align*}

```python
import numpy as np
import pandas as pd
import util 

from linear_model import LinearModel


def main_1b(train_path, eval_path, pred_path):
    """Problem 1(b): Logistic regression with Newton's Method.

    Args:
        train_path: Path to CSV file containing dataset for training.
        eval_path: Path to CSV file containing dataset for evaluation.
        pred_path: Path to save predictions.
    """
    # *** START CODE HERE ***
    x_train, y_train = util.load_dataset(train_path, add_intercept=True)
    x_eval, y_eval = util.load_dataset(eval_path, add_intercept=True)
    
    logreg = LogisticRegression()
    logreg.fit(x_train,y_train)
    
    y_pred = logreg.predict(x_eval)[1]
    
    # define data
    data = np.column_stack((x_eval[:,1:],y_eval,y_pred))
    df = pd.DataFrame(data, columns=['x_1','x_2','y','y_pred'])
    
    # save to csv file
    df.to_csv(pred_path, index=False, header=True, sep=',')
    
    # Output of the model 
    return (logreg.theta, np.mean(y_pred == y_eval))

    
    # *** END CODE HERE *** 
    
class LogisticRegression(LinearModel):
    """Logistic regression with Newton's Method as the solver.

    Example usage:
        > clf = LogisticRegression()
        > clf.fit(x_train, y_train)
        > clf.predict(x_eval)
    """

    def fit(self, x, y):
        """Run Newton's Method to minimize J(theta) for logistic regression.

        Args:
            x: Training example inputs. Shape (m, n).
            y: Training example labels. Shape (m,).
        """
        # *** START CODE HERE ***
        
        m, n = x.shape
        
        def h(theta,x): 
            return 1 / (1 + np.exp(-np.dot(x,theta)))
        
        def loss_function_gradient(theta,x,y): 
            return - 1 / m * np.dot(x.T, y-h(theta,x))
        
        def Hessian(theta,x): 
            return 1 / m * np.dot((h(theta,x) * (1-h(theta,x))) * x.T, x)
        
        def next_theta(theta,x,y): 
            return theta - np.dot(np.linalg.inv(Hessian(theta,x)),loss_function_gradient(theta,x,y))
        
        # Initial theta 
        if self.theta is None: 
            self.theta = np.zeros(n)
            
        # Find optimal theta
        old_theta = self.theta
        while self.eps >= 0.00001: 
            new_theta = next_theta(old_theta,x,y)
            self.eps = sum(abs(new_theta - old_theta))
            old_theta = new_theta

        self.theta = new_theta   
        # *** END CODE HERE ***

    def predict(self, x):
        """Make a prediction given new inputs x.

        Args:
            x: Inputs of shape (m, n).

        Returns:
            Outputs of shape (m,).
        """
        # *** START CODE HERE ***
        return (1 / (1 + np.exp(-(x @ self.theta))), 1 / (1 + np.exp(-(x @ self.theta))) >= 0.5)
        # *** END CODE HERE ***
```

```python
import numpy as np
import pandas as pd
import util

from linear_model import LinearModel


def main_1e(train_path, eval_path, pred_path):
    """Problem 1(e): Gaussian discriminant analysis (GDA)

    Args:
        train_path: Path to CSV file containing dataset for training.
        eval_path: Path to CSV file containing dataset for evaluation.
        pred_path: Path to save predictions.
    """
    # Load dataset
    x_train, y_train = util.load_dataset(train_path, add_intercept=False)

    # *** START CODE HERE ***
    x_eval, y_eval = util.load_dataset(eval_path, add_intercept=False)
    
    gda = GDA(LinearModel)
    gda.fit(x_train,y_train)
    
    y_pred = gda.predict(x_eval)[1]
    
    # define data
    data = np.column_stack((x_eval,y_eval,y_pred))
    df = pd.DataFrame(data, columns=['x_1','x_2','y','y_pred'])
    
    # save to csv file
    df.to_csv(pred_path, index=False, header=True, sep=',')
    
    # Accuracy of the model 
    return (gda.theta, gda.theta_0, np.mean(y_pred == y_eval))
    # *** END CODE HERE ***


class GDA(LinearModel):
    """Gaussian Discriminant Analysis.

    Example usage:
        > clf = GDA()
        > clf.fit(x_train, y_train)
        > clf.predict(x_eval)
    """

    def fit(self, x, y):
        """Fit a GDA model to training set given by x and y.

        Args:
            x: Training example inputs. Shape (m, n).
            y: Training example labels. Shape (m,).

        Returns:
            theta: GDA model parameters.
        """
        # *** START CODE HERE ***
        m, n = x.shape
        phi = np.sum(y) / m
        
        y_flip = np.where(y == 1, 0, 1)
        mu_0 = np.dot(x.T,y_flip)/np.sum(y_flip)
        mu_1 = np.dot(x.T,y)/np.sum(y)
        
        mu_x = np.dot(np.reshape(y,(m,1)),np.reshape(mu_1.T,(1,n))) + np.dot(np.reshape(y_flip,(m,1)),np.reshape(mu_0.T,(1,n))) 
        sigma = 1 / m * np.dot((x-mu_x).T, x-mu_x)
        sigma_inv = np.linalg.inv(sigma)
        
        self.theta = np.dot(sigma_inv,mu_1-mu_0)
        self.theta_0 = 1 / 2 * (mu_0.T @ sigma_inv @ mu_0 - mu_1.T @ sigma_inv @ mu_1) - np.log((1-phi) / phi)
        # *** END CODE HERE ***

    def predict(self, x):
        """Make a prediction given new inputs x.

        Args:
            x: Inputs of shape (m, n).

        Returns:
            Outputs of shape (m,).
        """
        # *** START CODE HERE ***
        return (x @ self.theta + self.theta_0, x @ self.theta + self.theta_0 >=0)
        
        # *** END CODE HERE
```

```python
import numpy as np
import util

def main_2e(train_path, valid_path, test_path, pred_path):
    """Problem 2: Logistic regression for incomplete, positive-only labels.
    Args:
        train_path: Path to CSV file containing training set.
        valid_path: Path to CSV file containing validation set.
        test_path: Path to CSV file containing test set.
        pred_path: Path to save predictions.
    """
    
    # *** START CODE HERE ***
    # Part (e): Apply correction factor using validation set and test on true labels
    x_train, y_train = util.load_dataset(train_path, label_col = 'y', add_intercept=True)
    x_eval, y_eval = util.load_dataset(valid_path, label_col = 'y', add_intercept=True)
    x_test, y_test = util.load_dataset(test_path, label_col = 'y', add_intercept=True)
    _, y_test_true = util.load_dataset(test_path, label_col = 't', add_intercept=True)
    
    logreg = LogisticRegression()
    logreg.fit(x_train,y_train)
    
    def h(theta,x): 
        return 1 / (1 + np.exp(-np.dot(x,theta)))
    
    x_V = x_eval[y_eval == 1]
    alpha = np.mean(h(logreg.theta,x_V))
    y_pred_raw = logreg.predict(x_test)[0]
    y_pred_raw_corrected = y_pred_raw / alpha 
    y_pred = np.where(y_pred_raw_corrected >= 0.5, 1, 0)

    # Output of the model 
    print("The accuracy using logistic regression (true labels) is: ", np.mean(y_pred == y_test_true))
    util.plot(x_test, y_test_true, logreg.theta, correction = alpha)
    
    # define data
    data = np.column_stack((x_test[:,1:],y_test, y_pred))
    df = pd.DataFrame(data, columns=['x_1','x_2','t', 'y_pred'])
    
    # save to csv file
    df.to_csv(pred_path, index=False, header=True, sep=',')
```

```python
# Dataset 3 (Q2e)

ds3_train_path = 'data/ds3_train.csv'
ds3_valid_path = 'data/ds3_valid.csv'
ds3_test_path = 'data/ds3_test.csv'
ds3_pred_path_e = 'data/ds3_pred_e.csv'

print("Logistic Regression using y labels with corrected alpha")
main_2e(ds3_train_path, ds3_valid_path, ds3_test_path, ds3_pred_path_e)
```

### 3. Poisson Regression

(a)
\begin{align*}
p(y;\lambda) 
&= \frac{e^{-\lambda}\lambda^y}{y!}\\
&= \frac{1}{y!}e^{(\log\lambda)y-\lambda}\\
\end{align*}

Thus, $b(y) = \frac{1}{y!}$, $T(y) = y$, $\eta = \log\lambda$, and $a(\eta) = \lambda = e^\eta$.

(b) $y|x\sim Pois(\lambda)$
\begin{equation*}
\mathbb{E}(y|x;\theta) = \lambda = e^\eta = e^{\theta^Tx}
\end{equation*}

(c) 
\begin{align*}
l(\theta) 
&= \log p(y^{(i)}|x^{(i)};\theta)\\
&= \log \frac{1}{y^{(i)}!}e^{y^{(i)\log\lambda}-\lambda}\\
&= \log\frac{1}{y^{(i)}!} + \log e^{y^{(i)}\log\lambda} + \log e^{-\lambda}\\
&= \log\frac{1}{y^{(i)}!} + y^{(i)}\log\lambda - \lambda\\
&= \log\frac{1}{y^{(i)}!} + y^{(i)}\theta^Tx^{(i)}-e^{\theta^Tx^{(i)}}
\end{align*}

Then, 

\begin{align*}
\frac{\partial l(\theta)}{\partial\theta_j} = (y^{(i)}-e^{\theta^Tx^{(i)}})x_j^{(i)}
\end{align*}

for all $j$. 

The update rule is 
\begin{align*}
\theta:=\theta + \alpha(y^{(i)}-e^{\theta^Tx^{(i)}})x^{(i)}
\end{align*}

repeated $m$ times. 

(d)


```python
import numpy as np
import util

from linear_model import LinearModel


def main_3d(lr, train_path, eval_path, pred_path):
    """Problem 3(d): Poisson regression with gradient ascent.

    Args:
        lr: Learning rate for gradient ascent.
        train_path: Path to CSV file containing dataset for training.
        eval_path: Path to CSV file containing dataset for evaluation.
        pred_path: Path to save predictions.
    """
    # Load training set
    x_train, y_train = util.load_dataset(train_path, add_intercept=False)
    x_eval, y_eval = util.load_dataset(eval_path, add_intercept = False)
    
    # *** START CODE HERE ***
    pois = PoissonRegression(step_size = lr)
    pois.fit(x_train, y_train)
    
    y_pred = pois.predict(x_eval)

    # define data
    data = np.column_stack((x_eval,y_eval,y_pred))
    df = pd.DataFrame(data, columns=['x_1','x_2','x_3', 'x_4', 'y','y_pred'])
    
    # save to csv file
    df.to_csv(pred_path, index=False, header=True, sep=',')
    
    # mean square error 
    return np.sqrt(np.mean(abs(y_pred-y_eval)))
    
    # *** END CODE HERE ***


class PoissonRegression(LinearModel):
    """Poisson Regression.

    Example usage:
        > clf = PoissonRegression(step_size=lr)
        > clf.fit(x_train, y_train)
        > clf.predict(x_eval)
    """

    def fit(self, x, y):
        """Run gradient ascent to maximize likelihood for Poisson regression.

        Args:
            x: Training example inputs. Shape (m, n).
            y: Training example labels. Shape (m,).
        """
        # *** START CODE HERE ***
        m, n = x.shape
        
        self.theta = np.zeros(n)

        for i in range(m): 
            self.theta += self.step_size * (y[i]- np.exp(self.theta.T @ x.T[:,i])) * x.T[:,i]
        
        # *** END CODE HERE ***
        

    def predict(self, x):
        """Make a prediction given inputs x.

        Args:
            x: Inputs of shape (m, n).

        Returns:
            Floating-point prediction for each input, shape (m,).
        """
        # *** START CODE HERE ***
        return np.exp(x @ self.theta)
        # *** END CODE HERE ***
```

### 4. Convexity of Generalized Linear Models 

(a) 
\begin{align*}
\mathbb{E}(Y|X;\theta) 
&= \int yb(y)e^{\eta y-a(\eta)}dy\\
&= \int \frac{\partial}{\partial\eta}b(y)e^{\eta y-a(\eta)}dy + \int b(y)e^{\eta y-a(\eta)}\frac{\partial a}{\partial\eta}dy \\
&= \frac{\partial a}{\partial\eta}
\end{align*}

(b) Note that 
\begin{align*}
0 = \frac{\partial^2}{\partial\eta^2} \int p(y,\eta) 
&= \int\frac{\partial}{\partial\eta}\left(b(y)\exp(ny-a(\eta))(y-a'(\eta)\right)\\
&= \int\left(b(y)\exp(ny-a(\eta))(-a''(\eta))+b(y)(y-a'(\eta))\exp(ny-a(\eta))(y-a'(\eta))\right)
\end{align*}

where $a'(\eta) = \frac{\partial a}{\partial\eta}$ and $a''(\eta) = \frac{\partial^2 a}{\partial\eta^2}$. 

Then, 
\begin{align*}
a''(\eta) 
&= \int(y-a'(\eta))^2b(y)\exp(ny-a(\eta))dy\\
&= \int(y-\mathbb{E}(Y|X;\theta))^2b(y)\exp(ny-a(\eta))dy\\
&= \mathbb{V}(Y|X;\theta)
\end{align*}

(c) Assume there is only one example in the training set. WLOG, it can be easily generalized to $m$ examples using similar steps.

\begin{align*}
J(\theta) 
&= -\log p(y|x;\theta) \\
&= -\log b(y)\exp(\eta y-a(\eta))\\
&= -\log b(y) -\left(\eta y-a(\eta)\right)
\end{align*}

Thus, 

\begin{align*}
\frac{\partial J(\theta)}{\partial\theta_j} 
&= \frac{\partial J(\theta)}{\partial\eta}\frac{\partial\eta}{\partial\theta_j}\\
&= (a'(\eta)-y)x_j
\end{align*}

Then, 

\begin{align*}
\frac{\partial^2 J(\theta)}{\partial\theta_k\partial\theta_j}
&= \frac{\partial}{\partial\theta_k}(a'(\eta)-y)x_j\\
&= \frac{\partial}{\partial\eta}(a'(\eta)-y)x_j\frac{\partial\eta}{\partial\theta_k}\\
&= a''(\theta^Tx)x_jx_k
\end{align*}

Therefore, 

\begin{align*}
z^THz &= a''(\theta^Tx)
\begin{bmatrix}
z_0 & z_1 & \cdots z_n
\end{bmatrix}
\begin{bmatrix}
x_0x_0 & x_0x_1 & \cdots & x_0x_n \\
x_1x_0 & x_1^2 & \cdots & x_1x_n \\ 
\vdots & & & \vdots\\
x_nx_0 & x_nx_1 & \cdots & x_nx_n
\end{bmatrix}
\begin{bmatrix}
z_0 \\ 
z_1 \\ 
\vdots \\ 
z_n
\end{bmatrix}\\
&= a''(\theta^Tx)\sum_i\sum_j z_ix_ix_jz_j \\
&= a''(\theta^Tx)(x^Tz)^2\\
&= \mathbb{V}(Y|X;\theta)(x^Tz)^2\geq 0
\end{align*}

### 5. Locally weighted linear regression 

(a) (i) $W$ is such that 
\begin{align*}
W_{ij} =
\begin{cases}
\frac{w^{(i)}}{2} & \text{if } i = j \\
0 & \text{if } i\neq j
\end{cases}
\end{align*}

(a) (ii) 
\begin{align*}
0 = \nabla_\theta J(\theta) 
&= \nabla_\theta(X\theta)^T-y^T)W(X\theta-y)\\
&= \nabla_\theta(X\theta)^TW(X\theta)-\nabla_\theta(X\theta)^TWy-\nabla_\theta y^TWX\theta + \nabla_\theta y^TWy\\
&= \nabla_\theta(X\theta)^TW(X\theta)-2\nabla_\theta y^TWX\theta+\nabla_\theta y^TWy\\
&= \nabla_\theta\theta^T(X^TWX)\theta-2\nabla_\theta(y^TWX)\theta\\
&= 2X^TWX\theta-2\nabla_\theta\theta^T(X^TW^Ty)\\
&= 2X^TWX\theta-2X^TW^Ty
\end{align*}

which implies 
\begin{align*}
\theta = (X^TWX)^{-1}(X^TWy)
\end{align*}

(a) (iii)
\begin{align*}
l(\theta)
&= \sum_{i=1}^m\log\frac{1}{\sqrt{2\pi}\sigma^{(i)}}\exp\left(-\frac{(y^{(i)}-\theta^Tx^{(i)})^2}{2\sigma^{(i)^2}}\right)\\
&= \sum_{i=1}^m\log\frac{1}{\sqrt{2\pi}\sigma^{(i)}} -\sum_{i=1}^m\frac{(y^{(i)}-\theta^Tx^{(i)})^2}{2\sigma^{(i)^2}}\\
&= \sum_{i=1}^m\log\frac{1}{\sqrt{2\pi}\sigma^{(i)}} -\sum_{i=1}^m\frac{w^{(i)}(y^{(i)}-\theta^Tx^{(i)})^2}{2}\\
&= \sum_{i=1}^m\log\frac{1}{\sqrt{2\pi}\sigma^{(i)}} - J(\theta)
\end{align*}

where $w^{(i)} = \frac{1}{\sigma^{(i)^2}}$. 

To maximise $l(\theta)$ it is equivalent to minimise $J(\theta)$ on the RHS. Thus, finding maximum likelihood estimate of $\theta$ reduces to solving a weighted linear regression problem. 

(b)

```python
import matplotlib.pyplot as plt
import numpy as np
import util

from linear_model import LinearModel


def main_5b(tau, train_path, eval_path):
    """Problem 5(b): Locally weighted regression (LWR)

    Args:
        tau: Bandwidth parameter for LWR.
        train_path: Path to CSV file containing dataset for training.
        eval_path: Path to CSV file containing dataset for evaluation.
    """
    # Load training set
    x_train, y_train = util.load_dataset(train_path, add_intercept = True)
    x_eval, y_eval = util.load_dataset(eval_path, add_intercept = True)

    # *** START CODE HERE ***
    # Fit a LWR model
    loc_weight = LocallyWeightedLinearRegression(tau)
    loc_weight.fit(x_train,y_train)
    y_pred = loc_weight.predict(x_eval)
    
    # Get MSE value on the validation set
    mse = np.mean(np.square(y_pred - y_eval))

    # Plot validation predictions on top of training set
    plt.scatter(x_eval[:,1],y_eval, marker = "x")
    plt.scatter(x_eval[:,1],y_pred, c = "r", marker = "o")
    plt.show()
    
    # No need to save predictions
    # Plot data
    
    return (mse,y_pred)
    # *** END CODE HERE ***


class LocallyWeightedLinearRegression(LinearModel):
    """Locally Weighted Regression (LWR).

    Example usage:
        > clf = LocallyWeightedLinearRegression(tau)
        > clf.fit(x_train, y_train)
        > clf.predict(x_eval)
    """

    def __init__(self, tau):
        super(LocallyWeightedLinearRegression, self).__init__()
        self.tau = tau
        self.x = None
        self.y = None

    def fit(self, x, y):
        """Fit LWR by saving the training set.

        """
        # *** START CODE HERE ***
        self.x = x 
        self.y = y 
        # *** END CODE HERE ***

    def predict(self, x):
        """Make predictions given inputs x.

        Args:
            x: Inputs of shape (m, n).

        Returns:
            Outputs of shape (m,).
        """
        
        # *** START CODE HERE ***
        def weightmatrix(item): 
            m1 = self.x.shape[0]
            w = np.zeros(m1)
            for i in range(m1): 
                w[i] = np.exp(-(np.sum((self.x[i:i+1,:] - item)**2)) / (2 * self.tau**2))
            W = np.eye(m1)
            np.fill_diagonal(W, w)
            return W
        
        m = x.shape[0]
        predict = np.zeros(m)
        
        for i in range(m): 
            theta_i = (np.linalg.inv(self.x.T @ weightmatrix(x[i:i+1,:]) @ self.x) @ self.x.T @ weightmatrix(x[i:i+1,:]) @ self.y)
            predict[i] = theta_i.T @ x[i:i+1,:].T
            
        return predict

        # *** END CODE HERE ***
```

```python
import matplotlib.pyplot as plt
import numpy as np
import util

def main(tau_values, train_path, valid_path, test_path, pred_path):
    """Problem 5(b): Tune the bandwidth paramater tau for LWR.

    Args:
        tau_values: List of tau values to try.
        train_path: Path to CSV file containing training set.
        valid_path: Path to CSV file containing validation set.
        test_path: Path to CSV file containing test set.
        pred_path: Path to save predictions.
    """
    # *** START CODE HERE ***
    # Search tau_values for the best tau (lowest MSE on the validation set)
    best = 10000
    best_tau = 10000
    for tau in tau_values: 
        MSE = main_5b(tau, train_path, valid_path)[0]
        print("The MSE when tau = ", tau, " is: ", MSE)
        if MSE < best: 
            best = MSE
            best_tau = tau
    
            
    print("Conclusion: The best tau is: ", best_tau)
    
    # Run on the test set to get the MSE value and Plot the data 
    print("With this best tau, the MSE on testing data is :", main_5b(best_tau, ds5_train_path, ds5_test_path)[0])
    
    # Save predictions to pred_path
    # define data
    x_test, y_test = util.load_dataset(test_path, add_intercept=True)
    y_pred = main_5b(best_tau, train_path, test_path)[1]
    data = np.column_stack((x_test[:,1],y_test, y_pred))
    df = pd.DataFrame(data, columns=['x_1','y', 'y_pred'])
    
    # save to csv file
    df.to_csv(pred_path, index=False, header=True, sep=',')
    
    # *** END CODE HERE ***
```

```python
# Dataset 5

ds5_train_path = 'data/ds5_train.csv'
ds5_valid_path = 'data/ds5_valid.csv'
ds5_test_path = 'data/ds5_test.csv'
ds5_pred_path = 'data/ds5_pred.csv'
tau_values = [3e-2, 5e-2, 1e-1, 5e-1, 1e0, 1e1]

main(tau_values, ds5_train_path, ds5_valid_path, ds5_test_path, ds5_pred_path)
```
